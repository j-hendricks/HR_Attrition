```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv("Resources/HR-Employee-Attrition.csv")

df.head()
```

```python
df_cat = df.dtypes[df.dtypes == 'object'].index.tolist()
```

```python
df[df_cat].nunique()
```

```python
enc = OneHotEncoder(sparse=False)

encode_df = pd.DataFrame(enc.fit_transform(df[df_cat]))

encode_df.columns = enc.get_feature_names(df_cat)

encode_df.head()
```

```python
df = df.merge(encode_df, left_index=True, right_index=True)

df = df.drop(df_cat, 1)

df.head()
```

```python
y = df['Attrition_Yes'].values

X = df.drop(['Attrition_Yes','Attrition_No'], 1).values

X_train, X_test, y_train, y_test = train_test_split(y, X)
```

```python
scaler = StandardScaler()

X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
num_input_features = len(X_train[0])
hidden_nodes1 = 8
hidden_nodes2 = 5

nn = tf.keras.models.Sequential([
    
    tf.keras.layers.Dense(units = hidden_nodes1, input_dim= num_input_features, activation='relu'),
    tf.keras.layers.Dense(units= hidden_nodes2, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
    
])

nn.summary()
```

```python
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
