```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
from collections import Counter
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Resources/HR-Employee-Attrition.csv")

df.head()
```

```python
df.columns.unique()
```

```python
df_cat = df.dtypes[df.dtypes == 'object'].index.tolist()
df_cat
```

```python
df[df_cat].nunique()
```

```python
enc = OneHotEncoder(sparse=False)

encode_df = pd.DataFrame(enc.fit_transform(df[df_cat]))

encode_df.columns = enc.get_feature_names(df_cat)

encode_df.head()
```

```python
df = df.merge(encode_df, left_index=True, right_index=True)

df = df.drop(df_cat, 1)

df.head()
```

```python
y = df['Attrition_Yes'].values

X = df.drop(['Attrition_Yes','Attrition_No'], 1).values

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
print(Counter(y_train))
print(Counter(y_test))
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
num_input_features = len(X_train[0])
hidden_nodes1 = 8
hidden_nodes2 = 5

# nn = tf.keras.models.Sequential([
    
#     tf.keras.layers.Dense(units = hidden_nodes1, input_dim= num_input_features, activation='relu'),
#     tf.keras.layers.Dense(units= hidden_nodes2, activation='relu'),
#     tf.keras.layers.Dense(units=1, activation='sigmoid')
    
# ])

# nn.summary()

input_tensor = tf.keras.layers.Input(shape=(num_input_features,))

x1 = tf.keras.layers.Dense(hidden_nodes1, activation='relu')(input_tensor)
x2 = tf.keras.layers.Dense(hidden_nodes2, activation='relu')(x1)

output_tensor = tf.keras.layers.Dense(1, activation='sigmoid')(x2)

model = tf.keras.models.Model(input_tensor, output_tensor)

model.summary()
```

```python
plot_model(model)
```

```python
model.compile(optimizer= 'adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
fit_model = model.fit(X_train, y_train, epochs=100, validation_data = (X_test, y_test))
```

```python
acc = fit_model.history['accuracy']
val_acc = fit_model.history['val_accuracy']
loss = fit_model.history['loss']
val_loss = fit_model.history['val_loss']
```

```python
plt.plot(acc, label='accuracy')
plt.plot(val_acc, label='val_accuracy')
plt.legend()
plt.title("History")
plt.show()
```

```python
plt.plot(loss, label='loss')
plt.plot(val_loss, label='val_loss')
plt.title("History")
plt.legend()
plt.show()
```
